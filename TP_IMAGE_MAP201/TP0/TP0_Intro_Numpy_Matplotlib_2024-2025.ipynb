Python propose une syntaxe qui permet de cacher les boucles for, c'est à dire tous les traitements itératifs.

Mais :

Voir aussi : https://realpython.com/python-for-loop/

```python
# Exemple de boucle deja vu plus haut
for i in range(5):
    print(i)  # NB: la partie du code répétée doit être indentée (1 tab ->)
```

```python
# Double boucle donc double indentation
for i in range(3):
    for j in range(2):
        print(f'Ici, i={i} et j={j}')

# Notez l'ordre d'affichage des indices

# i est dit 'least varying index' ou 'indice le plus externe'
# j est dit 'most varying index' ou 'indice le plus interne'
```

```python
# La boucle interieure peut dependre de la boucle exterieure
for i in range(3):
    for j in range(i+1):
        print(f'Et maintenant, i={i} et j={j}')
```

```python
# Boucle retrograde avec un pas de -1
# Commence a i=5 et s'arrete lorsque i=-1 donc la derniere valeur de i est i=0
for i in range(5, -1, -1):
    print(f'Retro loop: {i}')  # --> 5, 4, 3, 2, 1, 0
```

```python
# Boucle retrograde avec un pas de -2
# Commence a i=6 et s'arrete lorsque i=-1 donc la derniere valeur de i est i=0
for i in range(6, -1, -2):
    print(f'Retro loop: {i}')  # --> 6, 4, 2, 0
```

### Tests et boolens

Les tests et les booléens sont essentiels en Python pour écrire des programmes logiques et prendre des décisions. Voici un guide complet pour comprendre ces concepts.

#### Les booléens
En Python, il y a deux valeurs booléennes:
- `True` (vrai)
- `False` (faux)

```python
a = True
b = False
print(a)  # Affiche: True
print(b)  # Affiche: False
```

Les opérateurs de comparaison:

| Opérateur | Description                  | Exemple           |
|-----------|------------------------------|-------------------|
| `==`      | Égal à                       | `5 == 5` → `True` |
| `!=`      | Différent de                 | `5 != 3` → `True` |
| `<`       | Strictement inférieur à      | `3 < 5` → `True`  |
| `>`       | Strictement supérieur à      | `5 > 3` → `True`  |
| `<=`      | Inférieur ou égal à          | `3 <= 3` → `True` |
| `>=`      | Supérieur ou égal à          | `5 >= 3` → `True` |

Les opérateurs logiques:

| Opérateur | Description                  | Exemple                        |
|-----------|------------------------------|--------------------------------|
| `and`     | Tous les tests doivent être vrais | `(5 > 3) and (2 < 4)` → `True` |
| `or`      | Au moins un test doit être vrai   | `(5 > 3) or (2 > 4)` → `True`  |
| `not`     | Inverse le résultat d’un test    | `not(5 > 3)` → `False` |

```python
a = 10
b = 5
c = 15

print((a > b) and (a < c))  # True : 10 est supérieur à 5 ET inférieur à 15
print((a > b) or (a > c))   # True : 10 est supérieur à 5 OU supérieur à 15
print(not(a > b))           # False : L'inverse de True est False

```

#### Les tests `if`, `elif`, `else`.
Les tests sont souvent utilisés dans des instructions conditionnelles pour exécuter du code en fonction de certaines conditions.



```python
x = 10

if x > 5:
    print("x est supérieur à 5")
elif x == 5:
    print("x est égal à 5")
else:
    print("x est inférieur à 5")
```

Certaines valeurs sont automatiquement considérées comme True ou False dans un contexte booléen :

`False` : `None`, `0`, chaînes vides (`""`), listes vides (`[]`), dictionnaires vides (`{}`), etc.
`True` : Toutes les autres valeurs.

Examinez attentivement le résultat des instructions suivantes pour bien comprendre.

### Fonctions

```python
# Une fonction peut prendre des parametres (ou pas)
# et peut renvoyer une valeur (ou pas)

# Aucun paramètre, aucune valeur de retour
def no_param_no_return():
    print('Je ne prends aucun param et je ne renvoie rien')


# 3 paramètres dont un optionnel (noter la valeur par défaut), et une valeur de retour
def calcul_savant(x, y, z_optionnel=0):
    ln_x = math.log(x)
    sqrt_y = math.sqrt(y)
    return ln_x*sqrt_y + z_optionnel

# Appel de la fonction sans paramètres
no_param_no_return()

# Appel de la fonction avec paramètres
print(f'Resultat 1= {calcul_savant(42, 3.14)}')
print(f'Resultat 2= {calcul_savant(42, 3.14, 1000)}')

# Les valeurs données aux paramètres s'appelent des 'arguments' (ou args)
```

```python
# En python une fonction peut meme renvoyer plusieurs resultats!
def euclidean_div(x, y):
    # On suppose que y != 0
    reste = x % y
    quotient = (x - reste) / y
    return int(quotient), reste
    # Le quotient est converti en entier par l'instruction int(...)


# On récupére les résultats multiples avec la syntaxe suivante
q, r = euclidean_div(17, 5)
print(f'17/5 = {q}, reste {r}')

# Si on souhaite ignorer un résultat, une convention possible est de
# l'assigner à une variable 'poubelle' notée '_'
a = 100
b = 99
_, r = euclidean_div(a, b)
print(f'{a} modulo {b} = {r}')
```

### Les tableaux NumPy

#### Les tableaux 1D

```python
x = np.array([1,2,7,9])
print(x)
print(type(x))
```

Ne pas confondre un tableau 1D (`numpy`) et une liste (`Python`): un tableau 1D est homogène (il ne contient des éléments que d'un seul type). Les tableaux `numpy` sont très optimisés en termes de mémoire et sont bien mieux adaptées au calcul scientifique. On n'utilisera plus que des tabeaux `numpy` dans ce cours.

```python
import numpy as np
import time

# Taille des données
taille = 10**6

# Création des données
liste1 = list(range(taille))
liste2 = list(range(taille))

array1 = np.arange(taille)
array2 = np.arange(taille)

# Somme des listes Python
debut_liste = time.time()
resultat_liste = [x + y for x, y in zip(liste1, liste2)]
fin_liste = time.time()

# Somme des arrays NumPy
debut_numpy = time.time()
resultat_numpy = array1 + array2
fin_numpy = time.time()

# Affichage des temps d'exécution
print(f"Temps d'exécution (Listes Python) : {fin_liste - debut_liste:.5f} secondes")
print(f"Temps d'exécution (Arrays NumPy)  : {fin_numpy - debut_numpy:.5f} secondes")

```

#### Tableau 2D

```python
y = np.array([[1,2,3],[7,8,9]])
print(y)
print(type(y))
print(y.shape)
print(y.dtype)

# Note the differences
z = np.array([[1.,2,3],[7,8,9]])
print(z)
print(type(z))
print(z.shape)
print(z.dtype)
print(z.size)
```

Un tableau numpy ne peut contenir que des éléments d'un même type.
Vous manipulerez essentiellement des tableaux de **nombres**.
Les types de nombres principaux dans Python sont:

| Type     | Description                    | Plage de valeurs                          |
|----------|--------------------------------|-------------------------------------------|
| `np.uint8`  | Entier sur 8 bits (non signé)  | De 0 à 255                                |
| `np.uint16` | Entier sur 16 bits (non signé) | De 0 à 65,535                             |
| `np.uint32` | Entier sur 32 bits (non signé) | De 0 à 4,294,967,295                      |
| `np.uint64` | Entier sur 64 bits (non signé) | De 0 à 18,446,744,073,709,551,615         |
| `np.float16`   | Virgule flottante sur 16 bits     | Environ 3 décimales        |
| `np.float32`   | Virgule flottante sur 32 bits     | Environ 6-7 décimales      |
| `np.float64`   | Virgule flottante sur 64 bits     | Environ 15-16 décimales    |

```python
# Le type est choisi par numpy en fonction des valeurs du tableau.
x = np.array([1,2,3])
print(type(x[0]))

x = np.array([1.,2,3])
print(type(x[0]))

# Mais l'utilisateur peut aussi choisir le type après la création du tableau
y = x.astype(np.uint8)
print(type(y[0]))
print(y)

# ou à la création
z = np.array([1,4,8], dtype=np.float32)
print(type(z[0]))
```

Le choix du type repose sur un compromis entre:
- La précision : un calcul scientifique précis requiert généralement des `np.float32`, voire `np.float64`dans de nombreux cas.
- L'espace mémoire occuppé : un entier `np.uint8` occuppe 8 fois moins de place qu'un `np.float64`.

#### Comment accéder/modifier aux éléments d'un tableau
Observer les différentes commandes suivantes pour comprendre le concept de `slicing`, clé dans `numpy`.


```python
# Dans un tableau 2D, on accède aux éléments du tableau par un couple de coordonnées (num_ligne, num_col). Attention: l'ordre des coordonnées (i,j) n'est pas celui dont vous avez l'habitude (x,y). Les coordonnées commencent à ZERO. Ainsi, si un tableau à 10 éléments, l'indice 0 est celui du premier élément, l'indice 9 est celui du dernier.

x = np.arange(30).reshape((5,6))
print(f"x={x}")
print(x[0,4],x[4,5])
print(x[0], x[:,0])
print(x[1:3,4])
print(x[-1,-3:-1])
print(x[::2,1::2])

# Notez la syntaxe sur chaque dimension: start:end:step, où chaque terme est facultatif et peut être négatif.
# Bien comprendre cette syntaxe, centrale dans numpy.
print(np.arange(10)[::-1])
print(np.arange(10)[:-3])
```

Quelques éléments à garder en tête:
- Les indices de lignes, colonnes... démarrent à 0.

- Plus généralement (et plus compact), on peut directement entrer `x[-1]`, `x[-2]`... pour obtenir le dernier, l'avant-dernier... éléments de la liste `x`.

#### Fonctions statistiques

```python
# Un tableau peut être vu comme une série (de séries) statistique(s).ipynb_checkpoints.

x = np.random.rand(1000)
print(x.shape)
print(x.mean())  # La moyenne.
print(x.std())   # L'écart-type (le carré de la variance)
print(x.std()**2) # La variance.
```

```python
# Lorsqu'on travaille avec des tableaux 2D, il faudra préciser sur quel(s) axe(s) vous souhaitez calculer des indicateurs statistiques.

x = np.vstack([k*np.ones(5) for k in range(5)])
print(x)

print(x.mean())
print(x.mean(axis=0)) # Moyenne des colonnes.
print(x.mean(axis=1)) # Moyenne des lignes.
```

#### Fonctions aléatoires

Numpy fournit de nombreuses fonctions permettant de générer des (tableaux de) nombres aléatoires.

```python
import numpy as np

# Genere de manière équiprobable un nombre entier entre 0 et 2, donc 0, 1 ou 2
for i in range(5):
    print(np.random.randint(3))  # Un nombre ENTIER pris au hasard entre 0 (inclus) et 3 (exclu).

print()

# On peut également générer des nombres réels entre 0.0 et 1.0, issus d'une loi uniforme sur [0,1]
for i in range(5):
    print(np.random.rand())  # La probabilité d'obtenir exactement 0.0 ou 1.0 est ... nulle (ou presque).
# Question: Comment générer des nombres aléatoires d'une loi uniforme entre 2 et 5.

# Et plus généralement, on peut générer des nombres aléatoires issus d'autres loi de probabilité
for i in range(5):
    print(np.random.randn())  # Pour une loi normale centrée réduite
    print(np.random.poisson(100)) # Pour une loi de poisson de paramètre 100

# Question: Comment générer des nombres aléatoires d'une loi normale de moyenne 2 et d'écart-type 5.
```

Il est également très facile de générer des tableaux (matrices) de valeurs aléatoires.

```python
import numpy as np

# Vecteur de nombres entiers aléatoires
x = np.random.randint(10, size=(3))
print(x)
print()

# Tableau de nombres entiers aléatoires
y = np.random.randint(10, size=(3,10))
print(y)
print()

# Vecteur de nombres réels aléatories (issus d'une loi uniforme entre 0 et 1)
z = np.random.rand(5)
print(z)
print()

# Tableau de nombres réels aléatoires issus d'une loi normale centrée réduite.
w = np.random.randn(5, 2)
print(w)
```

```python
# Numpy affiche les matrices de manière plus naturelle que des listes
# Python a 2 dimensions

# Exemple d'une liste Python 2d
a = [[1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]]
print(f'Liste 2d=\n{a}')
print()

# Avec numpy
npa = np.array(a)
print(f'Array 2d=\n{npa}')
# Ici, le saut à la ligne \n permet d'aller a la ligne pour conserver
# l'alignement des cellules
print()

# Le formattage se fait bien meme avec des valeurs flottantes
cos_npa = np.cos(npa)
print(f'Cos of array 2d=\n{cos_npa}')
```

En plus de `arange`, numpy fournit la fonction `linspace` qui permet d'échantillonner un intervalle de valeurs flottantes.

- Le 1er paramètre est le début de l'intervalle (atteint).
- Le 2ème paramètre est la fin de l'intercalle (atteint aussi, cette fois).
- Le 3ème paramètre est le nombre total de valeurs (en incluant les bornes de l'intervalle).

```python
import numpy as np

start = 1
stop = 10
nb_samples = 5
sp1 = np.linspace(start, stop, nb_samples)
print(f'{sp1} contient {len(sp1)} elements, que voici')
for i in sp1:
  print(i)

print()

# Et si on augmente le nombre d'échantillons
nb_samples = 11
sp2 = np.linspace(start, stop, nb_samples)
print(f'{sp2} contient {len(sp2)} elements, que voici')
for i in sp2:
  print(i)
```

### Courbe représentative de fonctions.

```python
# On utilise en fait le sous-module pyplot de matplotlib, qu'il faut d'abord importer.
# La convention usuelle est de l'aliaser 'plt'.
import matplotlib.pyplot as plt

# Un graphe est défini par les coordonnées x des points : 1ere liste
# et les coordonnées y des points : 2eme liste
x = np.linspace(-4,4,1000)
y = np.cos(x)

# On définit une figure avec subplots.
fig,ax = plt.subplots()
ax.plot(x,y,label='cos(x)')

```

```python
# On peut ajouter un autre graphe sur la meme figure, avec d'autres attributs graphique.
fig,ax = plt.subplots()
ax.plot(x,y,label='cos(x)')
ax.plot(x,np.sin(x),color='r',linewidth=3,label='sin(x)')
```

```python
# Les labels (étiquette) permettent d'afficher une légende.
fig,ax = plt.subplots()
ax.plot(x,y,label='cos(x)')
ax.plot(x,np.sin(x),color='r',linewidth=3,label='sin(x)')
ax.legend()
```

On peut aussi tracer les deux fonctions sur deux sous-figure d'une même figure, et utiliser les titres pour décrire les courbes.

```python
fig,ax = plt.subplots(1,2)
ax[0].plot(x,y)
ax[0].set_title('cos(x)')
ax[1].plot(x,np.sin(x))
ax[1].set_title('sin(x)')
```

### Les nuages de points (*scatter* plot)

```python
# Le *scatter plot* ne fait que tracer les points (sans chercher à les relier comme dans le cas précédent).

y = np.linspace(-1,1,500)
x1 = np.sqrt(1 - y**2)
x2 = - np.sqrt(1 - y**2)

fig,ax = plt.subplots()
ax.axis('equal')
ax.scatter(x1,y,label='part 1')
ax.scatter(x2,y,label='part 2')
ax.legend()

```

```python
# Pour mieux ccomprendre la différence entre plot et scatter
x = np.random.rand(30)
y = np.random.rand(30)
fig,ax = plt.subplots(1,2)
ax[0].plot(x,y)
ax[1].scatter(x,y)
```

#### L'histogramme

```python
# Matlplotlib permet aussi de représenter graphiquement une série statistique, par un histogramme. L'histogramme permet d'avoir une représentation graphique de la distribution statistique des valeurs de la série.

x = [1,3,3,2,5,6,3,2,1,6]
fig,ax = plt.subplots()
ax.hist(x) # Essayer ensuite d'ajouter le paramètre bins=1, 10, 100 et comprendre l'effet.

# Avec des séries statistiques plus conséquentes
taille_serie = 10000 # Nombre de points dans la série
y = np.random.rand(taille_serie)
z = np.random.randn(taille_serie)
fig,ax = plt.subplots(1,2)
_,_,_ = ax[0].hist(y,bins=50) # Les "_,_,_" permettent de cacher certaines valeurs retournées.
ax[0].set_title("Loi uniforme.")
_,_,_ = ax[1].hist(z,bins=50)
ax[1].set_title("Loi normale")
```

#### Un exemple complet.

```python
# Creer une figure qui contient deux graphes ('axes' dans matplotlib)
# 1, 2 : une ligne de 2 figures
# figsize : dimensions de la figure, en pouces
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))

# L'instruction ci-dessus renvoie une figure fig et deux
# graphes (axes) ax1 et ax2 (dans un tuple)

# Utiliser la figure pour ajuster l'espacement entre les graphes
# pour eviter d'eventuels empietements
fig.tight_layout(pad=5.0)

# -------------------------------
# Parametrer le premier graphe
# -------------------------------

# Titre
ax1.set_title('Left graph')

# Legendes des axes X et Y
ax1.set_xlabel('X_units')
ax1.set_ylabel('Y_units')

# Intervalles des axes X et Y
ax1.set_xlim(0, 100)
ax1.set_ylim(-5, 5)

# Calcul des données à afficher
x1 = np.arange(1, 100, 0.05)
y1 = np.log(x1)*np.cos(x1 / 2)

# Affichage et generation de la legende
ax1.plot(x1, y1, color='green', label='courbe 1')
ax1.legend()

# -------------------------------
# Parametrage le second graphe
# -------------------------------

# Titre
ax2.set_title('Right graph')

# Legendes des axes X et Y
ax2.set_xlabel('X(t)')
ax2.set_ylabel('Y(t)')

# Intervalles des axes X et Y
ax2.set_xlim(-80, +80)
ax2.set_ylim(-80, +80)

# Calcul des données à afficher
t = np.arange(1, 100, 0.05)
r = t
x2 = r*np.cos(t)
y2 = r*np.sin(t)

# Affichage et generation de la legende
ax2.plot(x2, y2, color='magenta', label='spirale')
ax2.legend()

# -------------------------------
# Enregistrement de la figure
# -------------------------------

fig.savefig('./_TP0_TEST_FigMatplotlib.png')
```
