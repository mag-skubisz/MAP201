## Exercice 2 - Méthode d'Euler explicite


On considère l'équation 
$$\dot{x}(t)=f(t,x(t)),\ x(0)=1.$$
où $f(t,x)=-x+\sin(t)$, et on cherche à approcher la solution $t\mapsto x(t)$ sur l'intervalle de temps $t\in [0,T^{fin}]$ où $T^{fin}=10$. On rappelle la méthode d'Euler explicite vue en cours: on fixe un grand entier $N$ (le nombre de pas), et on pose

$$t_i=\frac{i T^{fin}}{N},\ \Delta t=\frac{T^{fin}}{N}$$
et $(x_i)_{i=0,...,N}$ est la suite (finie) définie par récurrence
$$
\begin{cases}
x_0=x^{init}\\
x_{i+1}=x_i+\Delta t\ f(t_i,x_i),\ i=0,1,2,...,N-1
\end{cases}
$$

telle que $x_i$ est une ``bonne approximation'' de la valeur de la solution $t\mapsto x(t)$ au temps $t_i$.

### 1)

Écrire une fonction (appelée $eulerexp$) prenant en entrée un entier $N\in\N^*$, et qui renvoie les tableaux $T=[t_0,t_1,...,t_N],X=[x_0,x_1,...,x_N]$, construits par la méthode ci-dessus.

```python
#Réponse:
def f(t,x):
    return -x + np.sin(t)

def eulerexp(N):
    Tfin=10
    td = Tfin / N
    T=np.linspace(0,Tfin,N+1)
    X=np.zeros(N+1)
    X[0]=1
    for i in range (N):
        X[i+1] = X[i] + td * f(T[i], X[i])

    return T,X
```

### 2)
Afficher le graphe de $(T,X)$ obtenu pour $N=10,15,30,100$. On représentera les quatre sur un même graphe.

```python
#Réponse:
colors = ["gray", "black", "pink", "violet"]
plt.figure(figsize=(8,6))
N = [10,15,30,100]
for i, N in enumerate(N):
    T,X = eulerexp(N)
    plt.plot(T,X, linestyle="--", color=colors[i])


x=T
y=(3/2)*np.exp(-x)-(1/2)*np.cos(x)+(1/2)*np.sin(x)
plt.plot(x,y, linestyle="-", color="blue") 
```

## Exercice 3 - Modèle de chute libre

Lorsqu'un parachutiste est en chute libre, un bon modèle de sa vitesse de chute $v(t)$ est l'équation
$$\begin{cases}
m\dot{v}(t)=mg-kv(t)^2,\text{ pour tout }t\geq 0\\
v(0)=0
\end{cases}$$
où $m>0$ est sa masse, $g>0$ est l'intensité de la pesanteur, et $k>0$ un coefficient de résistance de l'air. La distance parcourue après un temps $t$, notée $d(t)$, suit l'équation
$$\begin{cases}

d(0)=0

### 3) 
Écrire une fonction qui prend en entrée un nombre de pas (dans la méthode d'Euler) $N$, et un paramètre $k>0$, et renvoie la distance totale parcourue par le parachutiste dans sa chute libre, calculée par méthode d'Euler en $N$ pas avec un coefficient de resistance $k$.

```python
#Réponse:

import numpy as np

def chute_libre_euler(N, k, m=118.4, g=9.8, T=116): #N pas, k - air, m - mass, g, T - temps
    h = T / N  # intervalle
    v = 0      # vitesse
    d = 0      # distance

    for _ in range(N):
        dv = (g - (k / m) * v**2) * h  # equation donne / m
        v += dv
        d += v * h

    return d #distance parcourue 

N = 100000
k = 0.18549
distance = chute_libre_euler(N, k)
print(f"Distance totale: {distance:.2f} m")
```

### 4)
Écrire une fonction qui prend une entrée les paramètres suivants: une fonction d'un paramètre réel $f(=f(x))$, deux nombres $a<b$ tels que $f(a)$ et $f(b)$ sont de signe opposés, $\epsilon>0$ un nombre, et qui renvoie en sortie un nombre $c$ tel que $a<c<b$ et $c$ est à distance $\epsilon$ d'une racine de $f$.

On pourra procéder par dichotomie grâce à l'algorithme suivant:
- 1  On pose $x\leftarrow a,y\leftarrow b$.
- 2  Si $f(x)f((x+y)/2)<0$, alors on pose $y\leftarrow(x+y)/2$, sinon on pose $x\leftarrow (x+y)/2$.
- 3  On répète l'étape $2$ tant que $|y-x|>\epsilon$.
- 4  On renvoie $(x+y)/2$.

```python
#Réponse:

def bisection_method(f, a, b, epsilon):
    x, y = a, b
    while abs(y - x) > epsilon:
        m = (x + y) / 2  
        if f(x) * f(m) < 0:
            y = m  #  [x, m]
        else:
            x = m  #  [m, y]

    return (x + y) / 2  # approximation

def f(x):
    return x**3 - x - 2

R = bisection_method(f, 1, 2, 1e-6)  # f - fonction, [1,2] - intervalle, 1e-6 - precision
print(f"Approximation de la racine: {R:.6f}")
```
